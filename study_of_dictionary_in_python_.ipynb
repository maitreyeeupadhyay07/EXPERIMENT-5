```python
#changing/adding key:value in dictionary
car ={
    "brand":"ford",
    "model":"mustang",
    "year":1964
}
car["color"]="red"
print(car)

x=car.values()
print(x) #after the change
print(type(car))
```

```python
# Retrival of values
student ={
    "roll":101,
    "name":"Amit",
    "branch":"E&TC"
}
student.pop("branch")
print("After removal:",student)
student.popitem()
print("After removal:",student)
```
